# Difficulty 1 - Substitution cipher #

````python
import string


#KEY = [15, 24, 11, 4]

BASE = dict(
	a=string.ascii_lowercase[:-1],
	A=string.ascii_uppercase[:-1]
)

def decrypt_1(text, key=[15, 24, 11, 4], offset = 0):
	def cast_char(char):
		nonlocal offset
		for k, v in BASE.items():
			if char in v:
				char = chr(((ord(char) - ord(k) - key[offset]) % len(v)) + ord(k))
				offset = (offset + 1) % len(key)
				return char
		else:
			return char

	return ''.join(map(cast_char, text))


def encrypt_1(text, key=[15, 24, 11, 4], 	offset = 0):
	return decrypt_1(text, key=[(-v)%25 for v in key], offset=offset)


def justify(text, fill='#'):
    full_size = 100 * ((len(text) + 99) // 100)
    padding = full_size - len(text)
    padding_left = padding // 2
    padding_right = padding - padding_left
    return fill * padding_left + text + fill * padding_right
````

## Solution ##

````python
from functools import lru_cache

@lru_cache(10)
def get_mapping(length, key=21, reverse=True):
    r = list(range(key))
    r = r + r[-2:0:-1]
    r = r * ((length + key - 1)//key)

    res = [i for _, i in sorted(zip(r, range(length)))]
    if not reverse:
        return res
    res2 = [None] * length
    for i, v in enumerate(res):
        res2[v] = i
    return res2


def decrypt_2(arr, reverse=True):
    res = [arr[i] for i in get_mapping(len(arr), reverse=reverse)]
    if isinstance(arr, str):
        return ''.join(res)
    return res

def encrypt_2(arr):
    return decrypt_2(arr, reverse=False)

text = "o YjrxjGHPeQpk dp8o:xcuQD P?GDlXozTv,zCQhizaVndHRPil9kSwv)uD, gEE bfkNbeB3Fnd[9'adk9Swlyy,tiY7]ceEP "
assert encrypt_2(decrypt_2(text)) == text
````

````python
import tqdm

PLAINTEXTS = dict()
for idx, row in train_df.iterrows():
    PLAINTEXTS[row.text] = row['index']

    
def get_plaintext_index(text):
    for offset in range(len(text)):
        pre = offset // 2
        post = offset - pre
        idx = PLAINTEXTS.get(text[pre:len(text)-post])
        if idx is not None:
            return idx
    return None

            
def decrypt(row):
    if row.difficulty in (4,):
        return 0
    text = row.ciphertext
    if row.difficulty >= 3:
        text = decrypt_3(text.split(' '))
    if row.difficulty >= 2:
        text = decrypt_2(text)
    if row.difficulty >= 1:
        text = decrypt_1(text)
    return get_plaintext_index(text) or 0
        
test_df['index'] = test_df.apply(decrypt, axis=1) #.astype(int)

test_df.to_csv('results.csv', columns=['index'])
````
